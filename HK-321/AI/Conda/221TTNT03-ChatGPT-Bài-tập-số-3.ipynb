# ADD THƯ VIỆN 

```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
import numpy as np
import random
from scipy.spatial.distance import cdist
np.random.seed(11)

%matplotlib inline

```

# Random dữ liệu 

```python
features, true_labels = make_blobs(n_samples=300,n_features=2,cluster_std=2.75,random_state=42)

kmeans = KMeans(n_clusters=4,random_state=0).fit(features)
color_dict = {0:'b', 1:'g', 2:'y',3:'k'}
colors = [color_dict[i] for i in kmeans.labels_]
plt.scatter(features[:,0],features[:,1], c = colors)
plt.show()
```

# Dùng Sklearn tìm centroid

```python
plt.scatter(features[:, 0], features[:, 1], c= colors, s=50)

centers = kmeans.cluster_centers_

plt.scatter(centers[:, 0], centers[:, 1], c='r', s=50)
plt.show()
print(centers)
```

# Hàm Tính Toán Tìm Tọa Độ Centroid

```python
rng = np.random.RandomState(4)
def find_clusters(X, n_clusters):
    i = rng.permutation(X.shape[0])[:n_clusters]

    #1. chọn cluster trong features
    centers = X[i]
    
    while True:
        # gắn nhẵn cho tâm gần nhất 
        labels = np.argmin(cdist(X, centers) ,axis = 1)
        
        # tìm centers mới và gắn nhẵn
        new_centers = np.array([X[labels == i].mean(0)
                                for i in range(n_clusters)])
        
        # kiểm tra điều kiện centers mới và cũ để dừng 
        if np.all(centers == new_centers):
            break
        centers = new_centers
    
    return centers, labels

```

```python
mediumCenters  = 10e6
final_labels =None
final_centers= None


for i in range(3):
    centers,labels = find_clusters(features,4)
    
    new_mediumCenters = np.argmin(cdist(features, centers))
    print(f'centers {i+1} : \n {new_mediumCenters}')
    
    if(new_mediumCenters < mediumCenters):
        mediumCenters = new_mediumCenters
        final_labels = labels
        final_centers= centers
        
    plt.scatter(features[:, 0], features[:, 1], c= labels, s=50)
    plt.scatter(centers[:, 0], centers[:, 1], c='r', s=50)
    plt.show()
    


print(f'giá trị nhỏ nhất centers \n {final_centers}')
plt.scatter(features[:, 0], features[:, 1], c= final_labels, s=50)
plt.scatter(final_centers[:, 0], final_centers[:, 1], c='r', s=50)
plt.show()
```

# Hàm Đoán Dữ Liệu

```python
def predict_point(kmeans,input_test):
    y_train = kmeans.predict(input_test)[0]

    df=features[kmeans.labels_==y_train]
    
    plt.scatter(df[:, 0], df[:, 1], c= 'b', s=50)
    plt.scatter(input_test[:,0],input_test[:,1], c = 'r' ,marker= 'H',s = 300)
    plt.show()
    return 'Nhóm : ' + str(y_train)
```

# Nhập Dữ Liệu Dự Đoán

```python
X = int(input('X : '))
Y = int(input('Y : '))

input_test = np.array([X, Y]).reshape(-1,2)

predict_point(kmeans,input_test)
```
